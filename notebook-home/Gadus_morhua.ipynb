## Data

The hearing capabilities of Cod were researched by Chapman and Hawkins (1973). The effective hearing threshold was estimated to be between 60Hz and 380 Hz. 

## Defining the filter values
Filter weights are computed by applying the M-weighting function described in Southall et al. (2007), appendix A, p. 500. Southall et al. developed this filter functio for marine mammals, so its aplication to fishes is not as it is intended. However, the M-wighting function was developed by analogy to the C-wighting function used to estimate loudness perception of very loud sounds by humans, so by analogy, this function might might yield a better hearing curve in fishes than just using audiogram data.

```r
# import the FFT scripts
source("FFT.R")

# filter frequencies from 20Hz to 24kHz
seq <- make.f.seq(0.02, 24)

# functional hearing from 60 Hz to 380 Hz
f.low = 0.06
f.high = 0.38

M.table <- make.M.table(seq, f.low, f.high)
M.table

plot.M(M.table, TRUE)
```

## Implement the filter
Implement a filter using the values from table above. Q is set to 1.414 (1 octave) for a smoother filter curve. The filtered sound is normalized at -3dB.

The generated command can be used to invoque sox by copy-pasting it in a Linux terminal.

```r
# sox command
inputfile <- "data/whitenoise.wav"
outputfile <- "data/M.ft_Gadus_morhua.mp3"
Q <- 1.414 # 1 octave
command <- sox_command(inputfile, outputfile, M.table, Q)
command
```
