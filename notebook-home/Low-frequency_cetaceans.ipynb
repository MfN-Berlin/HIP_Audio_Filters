Frequency-weighting for hypothetical group "low-range" cetaceans. Functional hearing range according to Southal et al. (2007), table 2, p 430. 

Use this filter when there is no audiogram data for one of these genera: *Balaena, Caperea, Eschrichtius, Megaptera, Balaenoptera*

```r
# import the FFT scripts
source("FFT.R")
```

```r
# filter frequencies from 20Hz to 24kHz
seq <- make.f.seq(0.02, 24)

# functional hearing from 7 Hz to 22 kHz
f.low = 0.007
f.high = 22

M.table <- make.M.table(seq, f.low, f.high)
M.table

plot.M(M.table, TRUE)
```

```r
# sox command
inputfile <- "data/whitenoise.wav"
outputfile <- "data/M.ft_low-frequency_cetaceans.mp3"
Q <- 1.414 # 1 octave
command <- sox_command(inputfile, outputfile, M.table, Q)
command
```
